```python
!pip install yfinance matplotlib --quiet

import yfinance as yf
import matplotlib.pyplot as plt
import datetime

# Aes offshore
tickers = ['AAPL', 'MSFT', 'NVDA']

# Definir hoje e ltimos 10 dias teis pra garantir cobertura
hoje = datetime.datetime.today().strftime('%Y-%m-%d')
dias_uteis = (datetime.datetime.today() - datetime.timedelta(days=14)).strftime('%Y-%m-%d')

# Baixar apenas dados recentes
data = yf.download(tickers, start=dias_uteis, end=hoje)[['Open', 'Close']]

# Plotar grfico separado por ao
for ticker in tickers:
    df = data.xs(ticker, level=1, axis=1)
    df[['Open', 'Close']].plot(title=f'{ticker} - Abertura e Fechamento (ltimos dias teis)')
    plt.xlabel('Data')
    plt.ylabel('Preo (USD)')
    plt.grid(True)
    plt.show()




```

```python
!pip install yfinance --quiet

import yfinance as yf
import pandas as pd
import datetime

tickers = ['AAPL', 'MSFT', 'NVDA']
hoje = datetime.datetime.today().strftime('%Y-%m-%d')
dias_uteis = (datetime.datetime.today() - datetime.timedelta(days=14)).strftime('%Y-%m-%d')

data = yf.download(tickers, start=dias_uteis, end=hoje)[['Open', 'Close']]

# Montar tabela
tabela = []
for ticker in tickers:
    df = data.xs(ticker, level=1, axis=1).copy()
    df.reset_index(inplace=True)
    df['Ativo'] = ticker
    tabela.append(df[['Date', 'Ativo', 'Open', 'Close']])

tabela_final = pd.concat(tabela).sort_values(by=['Date', 'Ativo']).reset_index(drop=True)
display(tabela_final)
```
